```python
import sys
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
from refnx.analysis import CurveFitter
from refnx.reflect import Slab
```

```python
sys.path.append('../')

from structureSE import RI
from dataSE import DataSE
from reflect_modelSE import ReflectModelSE
from objectiveSE import ObjectiveSE
```

```python
# dname = 'WVASE_PNIPAM_water.dat'
dname = 'W475-P(M-s-O)-8020-.75BB-water-rpt-3-20C-spectro_WVASEformat.dat'
data = DataSE(data=dname)
wavs = np.unique(data.wav)
print(wavs)
```

```python
si     = RI('../materials/silicon.csv', name='Silicon')
sio2   = RI('../materials/silica.csv',  name='Silica')
PNIPAM = RI('../materials/pnipam.csv',  name='PNIPAM')
water  = RI('../materials/water.csv',   name='Water')

PNIPAM_layer = PNIPAM(150)
PNIPAM_layer.thick.setp(vary=True, bounds=(10,300))

struc = water() | PNIPAM_layer | sio2(2.5) | si()

model = ReflectModelSE(struc, wavelength=658)

model.delOffset.setp(value=-3, vary=False, bounds=(-10,10))
model.VolumeFrac.setp(value=1, vary=True, bounds=(0,1.1))
```

# Fitting experimental Data

```python
fig, ax = plt.subplots()
axt = ax.twinx()

aois = [65]
psis = []
deltas = []

for dat in data:
    model.wav = dat._current_wav
    
    psi, delta = model(aois)
    psis.append(psi)
    deltas.append(delta)
     
    p = ax.scatter(dat.wav, dat.psi, color='r')
    d = axt.scatter(dat.wav, dat.delta, color='b')

    ax.legend(handles=[p,d], labels=['Psi', 'Delta'])
    ax.set(ylabel='Psi', xlabel='Wavelength, nm')
    axt.set(ylabel='Delta')

ax.plot(wavs, psis, color='r')
axt.plot(wavs, deltas, color='b')
```

```python
objective = ObjectiveSE(model, data)
```

```python
objective.residuals()
```

```python
fitter = CurveFitter(objective)
fitter.fit(method='differential_evolution');
```
